```python
import pandas as pd
```

```python
netflix_titles_df = pd.read_csv('MoviesOnStreamingPlatforms_updated.csv')
```

```python
Ages = netflix_titles_df["Age"]
Ages = Ages.str.rstrip('+')
Ages
```

```python
netflix_titles_df["Age"] = Ages
```

```python
netflix_titles_df['Rotten Tomatoes'] = netflix_titles_df['Rotten Tomatoes']/100 
```

```python
clean_df = netflix_titles_df.drop(columns={"Unnamed: 0","ID", "Hulu","Prime Video","Disney+","Type", "IMDb"})
clean_df
```

```python
clean_df = clean_df.dropna()
clean_df
```

```python
filt = clean_df["Netflix"] == 1
filt

clean_df = clean_df.loc[filt,:]
```

```python
clean_df.to_csv("Netflix_Rotten_Tomatoes_Scores.csv", index=False)
```
