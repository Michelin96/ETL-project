```python
import pandas as pd
```

```python
imdb_df = pd.read_csv('imdb.csv')
nf_df= pd.read_csv('Netflix_Rotten_Tomatoes_Scores.csv')
```

```python
imdb_df=imdb_df.rename(columns = {'primaryTitle':'Title', 'startYear':'Year', 'runtimeMinutes':'Runtime'})
```

```python
imdb_df.head()
```

```python
nf_df.head()
```

```python
final_merge= imdb_df.merge(nf_df, how='inner', on=('Title','Year','Runtime'))
```

```python
final_merge_columns = final_merge[['Title','Year','Runtime']]
final_merge_columns
```

```python
final_merge_columns.to_csv('junction_table.csv', encoding='utf-8', index=False)
```
